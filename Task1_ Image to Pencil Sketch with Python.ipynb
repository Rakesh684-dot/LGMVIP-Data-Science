## AIM- Image to Pencil Sketch with Python

### **IMPORTING ALL THE NECESSARY LIBRARIES**

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import sys
import os
```

### **LOADING THE IMAGE**

```python
image = cv2.imread('C:\\Users\\RAKESH\\Documents\\captain america.jpg')
```

### **PRINTING ORIGINAL IMAGE**

```python
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("ORIGINAL IMAGE")
plt.show()
```

### **PRINTING GRAYSCALE IMAGE**

```python
g_image=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
plt.imshow(cv2.cvtColor(g_image, cv2.COLOR_BGR2RGB))
plt.title("GRAYSCALE IMAGE")
plt.show()
```

### **PRINTING THE NEGATIVE IMAGE**

```python
invert=cv2.bitwise_not(g_image)
plt.imshow(cv2.cvtColor(invert, cv2.COLOR_BGR2RGB))
plt.title("NEGATIVE IMAGE")
plt.show()
```

### **PRINTING THE IMAGE IN PENCIL SKETCH FORM**

```python
blur=cv2.GaussianBlur(invert, (31,31), 0)
inv_blur=cv2.bitwise_not(blur)
sketch=cv2.divide(g_image,inv_blur, scale=256.0)
plt.imshow(cv2.cvtColor(sketch, cv2.COLOR_BGR2RGB))
plt.title("PENCIL SKETCH")
plt.show()
```
